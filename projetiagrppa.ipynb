```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tqdm import tqdm  # Pour afficher une barre de progression

# DÃ©finir le chemin des fichiers
file_paths = {
    "employee_survey_data": "employee_survey_data.csv",
    "general_data": "general_data.csv",
    "in_time": "in_time.csv",
    "manager_survey_data": "manager_survey_data.csv",
    "merged_data": "merged_data.csv",
    "out_time": "out_time.csv"
}

# Dictionnaire pour stocker les DataFrames
dataframes = {}

# Charger et traiter les fichiers avec une barre de progression
print("\nðŸ“¥ Chargement des fichiers...\n")
for name, path in tqdm(file_paths.items(), desc="Progression"):
    if os.path.exists(path):
        try:
            df = pd.read_csv(path)
            
            # Supprimer la colonne Over18 si elle existe
            if 'Over18' in df.columns:
                df.drop(columns=['Over18'], inplace=True)
            
            # Stocker dans le dictionnaire
            dataframes[name] = df

        except Exception as e:
            print(f"\nâŒ Erreur lors du chargement de {name}: {e}")

# VÃ©rifier si des fichiers ont Ã©tÃ© chargÃ©s
if not dataframes:
    print("\nâš  Aucun fichier chargÃ©. VÃ©rifiez vos fichiers CSV.")
else:
    # CrÃ©er un rÃ©sumÃ© des fichiers
    summary_data = []
    for name, df in dataframes.items():
        total_values = df.count().sum()  # Nombre total de valeurs non nulles
        missing_percentage = df.isnull().mean().mean() * 100  # Pourcentage de valeurs manquantes
        summary_data.append({
            "Fichier": name,
            "Lignes": df.shape[0],
            "Colonnes": df.shape[1],
            "Valeurs manquantes (%)": round(missing_percentage, 2),
            "Valeurs totales": total_values
        })

    # Convertir en DataFrame
    summary_df = pd.DataFrame(summary_data)

    # VÃ©rifier si le DataFrame contient des donnÃ©es avant de l'afficher
    if not summary_df.empty:
        print("\nðŸ“Œ RÃ©sumÃ© des fichiers:\n")
        print(summary_df.to_string(index=False))  # Format propre sans index

        # Affichage des valeurs exactes par fichier et par colonne
        print("\nðŸ”Ž DÃ©tail des valeurs par colonne et par fichier:\n")
        for name, df in dataframes.items():
            print(f"\nðŸ“‚ **Fichier: {name}**")
            print("=" * 70)
            for column in df.columns:
                total_values = df[column].count()
                missing_percentage = (df[column].isnull().sum() / df.shape[0]) * 100
                print(f"ðŸ“Œ Colonne '{column}': {total_values} valeurs non nulles ({round(missing_percentage, 2)}% manquantes)")
            print("=" * 70)

        # Graphique des valeurs manquantes par fichier
        plt.figure(figsize=(12, 6))
        sns.barplot(x=summary_df["Fichier"], y=summary_df["Valeurs manquantes (%)"], palette="coolwarm")
        plt.xlabel("Fichier")
        plt.ylabel("Valeurs manquantes (%)")
        plt.title("ðŸ“‰ Pourcentage de valeurs manquantes par fichier")
        plt.xticks(rotation=45)
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        # Afficher le graphique
        plt.show()
    else:
        print("\nâš  Aucun rÃ©sumÃ© disponible, aucun fichier chargÃ© correctement.")

print("\nâœ… Analyse terminÃ©e avec succÃ¨s !")
```
